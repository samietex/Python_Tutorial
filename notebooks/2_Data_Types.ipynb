# 1. BOOLEANS

A Boolean data type can have one of two values: `True` or `False`. Booleans are fundamental for conditional statements and logic in programming.

````python
Uzi = 5
David = 5

# Comparison operators

# == is equal to
# > is greater than
# < is less than
# >= is greater or equal to
# <= is less than or equal to
# != is not equal to

Uzi != David
````

**Example**

````python
# Boolean variables
is_python_fun = True
is_learning_python = False

# Logical operations
result_and = is_python_fun and is_learning_python
result_or = is_python_fun or is_learning_python
result_not = not is_python_fun

print(result_and)  # False
print(result_or)   # True
print(result_not)  # False
````

From the above lines of code, you can see that two variables are created. A boolean variable can only have two values: `True` and `False`. In this case, `is_python_fun` is assigned the value `True`, indicating that Python is fun, and `is_learning_python` is assigned the value `False`, indicating that the person is not currently learning Python.

In the logical operations section, three logical operations are performed using these boolean variables. Let's look at them:

* `result_and = is_python_fun and is_learning_python`: This line uses the logical AND (`and`) operator. The result (`result_and`) will be `True` only if both `is_python_fun` and `is_learning_python` are `True`. Otherwise, it will be `False`.

* `result_or = is_python_fun or is_learning_python`: This line uses the logical OR (`or`) operator. The result (`result_or`) will be `True` if at least one of `is_python_fun` or `is_learning_python` is `True`. It will be `False` only if both are `False`.

* `result_not = not is_python_fun`: This line uses the logical NOT (`not`) operator. The result (`result_not`) will be the opposite of the value of `is_python_fun`. If `is_python_fun` is `True`, then `result_not` will be `False`, and if `is_python_fun` is `False`, then `result_not` will be `True`.

# 2. LIST

List is a collection of items. These items can be of any data type (numbers, strings, other lists, etc.), and they are enclosed in square brackets []. Lists are mutable, meaning you can change, add, or remove elements after the list is created.

To create a list, you simply place the items inside square brackets, separated by commas.

````python
x = [2, 5, 6, 10]

x
````

````python
# A list is a collection which is ordered and changeable.
# List allows duplicate members.
datascience = [20, 40, 50, '3MTT', 'AI']
datascience
````

You can access individual elements in a list using their index. Python uses zero-based indexing, meaning the first element has an index of 0.

````python
datascience[0]
````

Please take note that indexing can either be positive indexing or negative indexing. Positive indexing starts numbering from the very first element in the list which has an index of `0`. However, in the case of negative indexing, numbering starts from the last element in the list with an index of `-1`.

Here's a good example.

````python
datascience[-1]
````

````python
datascience[-2]
````

### Slicing

Slicing allows you to create a new list by extracting a portion of an existing list. The syntax is `list[start:stop]` where `start` is the inclusive index and `stop` is the exclusive index.

In simpler terms, `start` is the first element you are considering in the list, while for `stop`, it is not included in the new list but the number before it.

Take for instance you want to create list containing 40, 50, and '3MTT' from the datascience list, the new list will be ===> `datascience[start:stop]`. Where `start` is the index of `40`, and `stop` is the index of the element after '3MTT'

````python
new_list = datascience[1:4]
new_list
````

### List Methods

Python provides several built-in methods for manipulating lists:

* `append()`: Adds an element to the end of the list.
* `extend()`: Appends the element of another list to the end of the list.
* `insert()`: Inserts an element at a specific position.
* `remove()`: Removes the first occurence of a specific value.
* `pop()`: Removes and returns the element at a specific positon.
* `index()`: Returns the index of the first occurences of a value.
* `count()`: Returns the number of occurences of a value.
* `sort()`: Sorts the elements of the list in ascending order.
* `reverse()`: Reverses the order of the elements in the list.

# 3. DICTIONARY

A dictionary is an unordered collection of items, where each item is a key-value pair. It allows you to store and retrieve values using a unique key. Dictionaries are defined using curly braces `{}`, and each key-value pair is separated by a colon `:`.

````python
# my_dict = {'key1':'value1', 'key2':'value2', 'key3':'value3'}
# Dictionary consists of a collection of key-value pairs. Each key-value pair maps the key to its corresponding value.
# Keys are unique within a dictionary while values may not be.
# List elements are accessed by their position in the list, via indexing while Dictionary elements are accessed via keys

# Define a dictionary using key-value pairs
students = {'name' : 'Oladimeji', 'names' : 'Tola', 'school' : 'Gomycode', 'age' : 10}
students
````

From the above line of code, the `students` variable is a dictionary with `name` as the first key and `Oladimeji` as the first value assigned to the first key.

You can access the the value associated with a specific key, using the following approach below:

# 6. SETS

A set is an unordered collection of unique elements. Unlike lists or tuples, sets do not have a specific order, and each element in a set must be unique. Sets are defined using curly braces `{}`.

````python
# A set is an unordered collection of items. Every element is unique (no duplicates).
# A set is created by placing all the items (elements) inside curly braces {}, separated by comma or by using the built-in function set().
# set of integers
Data_science = {'Olakunle', 'Sapphire', 67, 78}
Data_science
````

````python
Data_science.add('David')
````

````python
Data_science.remove(67)
````

````python
del Data_science
````

````python
Data_science
````

````python
# set do not have duplicates
my_set = {'Data Science', 'ML', 'AI', 'AI', 'Game Dev'}
print(my_set)
````

Alternatively, you can use the `set()` constructor to create a set:

````python
new_set = set(['3MTT', 4, 6, 'ML'])
new_set
````
