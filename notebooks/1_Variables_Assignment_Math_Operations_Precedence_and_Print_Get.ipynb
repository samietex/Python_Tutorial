# Python As a Dynamically Typed Language

## Introduction

Python is a dynamically typed programming language, which means it handles variable assignments differently compared to statically typed languages like C, C++, and Java. Unlike these languages, Python does not require explicit variable declarations; instead, it relies on assignment statements.

## Static Typing vs. Dynamic Typing

In statically typed languages like C, C++, or Java, when you declare a variable, you set aside a specific area of memory for holding values of a predefined data type. This memory allocation is interpreted strictly according to the data type of the variable. For example, if you declare an integer variable, the allocated memory will be treated as an integer, and attempting to store any other data type in it will result in a compile-time error. This rigid type checking ensures that variables are used consistently with their declared types.

## Dynamic Typing in Python

Python, on the other hand, is dynamically typed. In Python, there is no need to explicitly declare a variable's data type. Instead, Python stores values at specific memory locations and binds variable names to these memory containers. It doesn't determine the variable's type until the code is executed at runtime.

This dynamic typing approach has several implications:

1. No Explicit Declaration: In Python, you do not need to declare the type of a variable before assigning a value to it. You can simply assign a value, and Python will infer the variable's type based on the assigned value.

2. Type Flexibility: Python variables can change their type during execution. For example, you can assign an integer value to a variable and later reassign it to hold a string, and Python will adapt to these changes without any compile-time errors.

3. Run-Time Type Checking: Python checks variable types at runtime, not at compile time. It allows you to mix different data types in your code, and any type-related errors will only be discovered when the code is executed.

````python
# Dynamic typing in action
x = 42      # x is an integer
x = "Hello" # x is now a string

# No compile-time errors, as Python adapts to the changing type of x
````

In summary, Python's dynamic typing simplifies variable assignment and allows for greater flexibility in handling data types, as it determines variable types at runtime rather than compile time. This flexibility makes Python a versatile and expressive language for a wide range of programming tasks.

````python
# Define a variable named AAPL and initialize it with 5 indicating that we own 5 stocks
# Let's assume that we bought an additional stock so now we have 6 stocks instead of 5
# We can increment AAPL by 1 stock as follows:
AAPL = 100
AAPL = AAPL + 10
AAPL
````

`AAPL = 100`: Here, you are assigning the value `100` to the variable `AAPL`. This line initializes `AAPL` with the value `100`.

`AAPL = AAPL + 10`: In this line, you are updating the value of `AAPL`. `AAPL + 10` calculates the sum of the current value of `AAPL` (which is `100`) and `10`. The result (`110`) is then assigned back to the variable `AAPL`. So, after this line, `AAPL` holds the value `110`.

`AAPL`: If you print the value of `AAPL` after the above operations, it will output `110`. This is the final value of the variable AAPL after the addition.

````python
AAPL = 100
AAPL -= 20
AAPL
````

`AAPL = 100`: Here, you are assigning the value `100` to the variable `AAPL`. This line initializes `AAPL` with the value `100`.

`AAPL -= 20`: In this line, you are updating the value of `AAPL` by subtracting `20` from its current value. The expression `AAPL -= 20` is a shorthand for `AAPL = AAPL - 20`. So, it subtracts `20` from the current value of `AAPL` (which is `100`) and assigns the result (`80`) back to the variable `AAPL`.

`AAPL`: If you print the value of `AAPL` after the above operations, it will output `80`. This is the final value of the variable `AAPL` after the subtraction.

````python
# Alternatively, we can increment the variable AAPL by 1 as follows:
AAPL = 10
AAPL += 10
AAPL
````

````python
# Let's assume that the price of AAPL stock is $260 and we currently have 5 AAPL stocks in our portfolio
# We can calculate the total portfolio value (account balance) as follows:
AAPL_count = 20
AAPL_price = 2000
account_balance = AAPL_count * AAPL_price
account_balance
````

`AAPL_count = 20`: Here, you are assigning the value `20` to the variable `AAPL_count`. This line initializes `AAPL_count` with the quantity of AAPL (Apple Inc.) stocks you have, which is `20`.

`AAPL_price = 2000`: This line assigns the value `2000` to the variable `AAPL_price`. It represents the price of one share of `AAPL` stock, which is assumed to be $2000 in this example.

`account_balance = AAPL_count * AAPL_price`: Here, you calculate the total account balance by multiplying the number of `AAPL` stocks (`AAPL_count`) with the price per stock (`AAPL_price`). The result (`20 * 2000 = 40000`) is assigned to the variable `account_balance`.

`account_balance`: If you print the value of `account_balance` after the above operations, it will output `40000`. This is the final account balance calculated based on the number of `AAPL` stocks and their price per share.

````python
# Let's assume you have $20000 USD in our account
# We want to buy AMZN stocks using the total amount
# AMZN stock is priced at $3116 each

account_balance = 20000
AMZN_price = 3116

AMZN_count = account_balance//AMZN_price
AMZN_count
````

`account_balance = 20000`: This line initializes the variable `account_balance` with a value of `20000`, representing the amount of money available in the account.

`AMZN_price = 3116`: Here, you set the variable `AMZN_price` to 3116, which represents the price of one share of AMZN (Amazon.com Inc.) stock.

`AMZN_count = account_balance // AMZN_price`: This line calculates the maximum number of whole shares of `AMZN` stock that can be bought with the given `account_balance`. The `//` operator performs integer division, giving you the maximum count of stocks without considering fractional shares.

`AMZN_count`: If you print the value of `AMZN_count` after the above operations, it will output the calculated count of whole shares of AMZN stock that can be purchased with the given account balance.

````python
# % returns the remainder

AMZN_count = account_balance % AMZN_price 
AMZN_count
````

**`AMZN_count = account_balance % AMZN_price`**: This line calculates the remainder when the `account_balance` is divided by the `AMZN_price`. The `%` operator is the modulus operator, and it returns the remainder of the division.

**`**AMZN_count**`**: If you print the value of `AMZN_count` after the above operation, it will output the remainder, which represents the amount of money left after buying as many whole shares of AMZN stock as possible with the given account balance. This is essentially the "change" or "remainder" in terms of the stock price.

# 3. ORDER OF OPERATIONS (PRECEDENCE)

Order of Operations, also known as precedence, refers to the rules that dictate the sequence in which different operations are performed in a mathematical expression. In Python, as in many programming languages, understanding and following the order of operations is crucial for correctly evaluating expressions.

The order of operations in Python is as follows:

1. **Parentheses** (): Expressions inside parentheses are evaluated first. If there are nested parentheses, the innermost expressions are evaluated first.

2. **Exponents** **: Exponentiation is the next operation. This involves raising a number to a power.

3. **Multiplication** *, **Division** /, **Floor Division** //, **Modulus** %: These operations are performed from left to right.

4. **Addition** +, **Subtraction** -: Addition and subtraction are performed from left to right.

When an expression involves multiple operators, Python follows these rules to determine the correct order of evaluation. For example:

````python
# add 3 and 4 and then multiply the answer by 5
# Note that parantheses have the highest precedence
(2*5) + 5
````

````python
# Multiplication has higher precedence compared to addition
1 * 2 + 3
````

````python
# Division has higher precedence compared to addition
16 / 4 + 25 / 5
````
